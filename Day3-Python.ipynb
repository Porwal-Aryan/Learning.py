```python
#infinite loop program to check whether your number is postive,neg or zero

while(True):
  num = int(input("enter any number "))
  if(num>0):
    print("number is positive")
  elif(num<0):
    print("number is negative")
  else:
    print("number is zero")
```

```python
# termination case for infinite loop program to check whether your number is
# postive,neg or zero

while(True):
  num_input = input("enter any number (or type 'quit' to exit): ")
  if num_input.lower() == 'quit':
    break
  try:
    num = int(num_input)
    if(num>0):
      print("number is positive")
    elif(num<0):
      print("number is negative")
    else:
      print("number is zero")
  except Exception as e:
    print(f"Invalid input. Please enter a number or 'quit'.")
    print(f"stacktrace exception: {e}")
```
