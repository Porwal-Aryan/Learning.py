```python
platform_a = {"Aryan", "Asmit", "Amit", "Abhiskek"}
platform_b = {"Rahul", "Aryan", "Rohit", "Rohan"}
print(platform_a)
print(platform_b)

#union:
all_friends = platform_a | platform_b
print(all_friends)

#intersection:
comman_friends = platform_a & platform_b
print(comman_friends)

#difference
unique_friends_a = platform_a - platform_b
unique_friends_b = platform_b - platform_a
print(unique_friends_a)
print(unique_friends_b)

#combine all friends:
all_unique_friends = platform_a ^ platform_b
print(all_unique_friends)
```
